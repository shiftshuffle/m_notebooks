$\sum_{n=1}^\infty \frac{1}{n \cdot f(n)}$ converge
donde $f:\mathbb{N} \to \mathbb{N}$ es una biyeccion

**nota**: (por notacion aqui $N=\{1,2,...\}$ aunque considero que 0 es natural

-------

Sea $A=\{n | \sqrt{n} \leqslant f(n) \}$     y sea    $B=\mathbb{N} \setminus A$

$\therefore$

$\sum\limits_{n\in N} \frac{1}{n \cdot f(n)} = \sum\limits_{n\in A} \frac{1}{n \cdot f(n)} + \sum\limits_{n\in B} \frac{1}{n \cdot f(n)} $ 


Ahora si 
 * $n \in A$:  
   
   >$\sqrt(n) \leqslant f(n)$   
   >
  >$\frac{1}{n\cdot f(n)} \leqslant \frac{1}{n\cdot \sqrt(n)}$  
  
  >$\sum\limits_{n\in A} \frac{1}{n \cdot f(n)} \leqslant \sum\limits_{n\in A} \frac{1}{n \cdot \sqrt(n)} \leqslant \sum\limits_{n\in \mathbb{N}} \frac{1}{n^{\frac{3}{2}}} \therefore converge $
    
      
        
        
 * $n \in B$:
 
      >Sea $n \in B(m)=B\cap [1,m]$
     
     >$n \leqslant m$
     
     >$f(n) < \sqrt(n) \leqslant \sqrt(m)$      
     
     >$\therefore     |B| < \sqrt(m)$      (f inyectiva)
     
     >Sea $\lfloor \sqrt(m) \rfloor = a$  
     
     >$B(m)=\{n_{1},n_{2}...n_{a} \}$
     
     >Ordenamos los elementos de $B(m)$ de manera que $\frac{1}{n_{1}} < \frac{1}{n_{2}} ... < \frac{1}{n_{a}} $
     
     >$\sum\limits_{n\in B(m)} \frac{1}{n \cdot f(n)} = \frac{1}{n_{1}f(n_{1})} + \frac{1}{n_{2}f(n_{2})} ... + \frac{1}{n_{a}f(n_{a})} \leqslant \frac{1}{1 f(n_{1})} + \frac{1}{2 f(n_{2})} ... + \frac{1}{a f(n_{a})} $
     
     >Ahora $\frac{1}{f(n_{1})} , \frac{1}{f(n_{2})} ... , \frac{1}{f(n_{a})}$ es una permutacion de $\frac{1}{1}, \frac{1}{2}... , \frac{1}{a} $
     >utilizando la [desigualdad de rearreglo](https://en.wikipedia.org/wiki/Rearrangement_inequality)
     
     >tenemos entonces $\frac{1}{1 f(n_{1})} + \frac{1}{2 f(n_{2})} ... + \frac{1}{a f(n_{a})} \leqslant \frac{1}{1^{2}}+ \frac{1}{2^{2}}... + \frac{1}{a^{2}} $
     
     > $\therefore$ cuando hacemos  $ m \to\infty$  $\sum\limits_{n\in B(m)} \frac{1}{n \cdot f(n)} \leqslant \sum\limits_{n\in \mathbb{N}} \frac{1}{n^{2}} = \frac{\pi^{2}}{6} $ y por tanto converge
